# Research Question 1: Remixes vs Original Songs Popularity

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
# Read in CSV file of top Spotify artists
df = pd.read_csv('../../Data/Processed/deezer_tracks_2000_2025.csv')
df.head()
df.shape
```

```python
# Standardize column names
df.columns = df.columns.str.lower()

# Helper function: extract the main title (remove remix/alternate info)
def get_base_title(title):
    return re.sub(r'[\(\[].*?[\)\]]', '', title).strip().lower()

df['main_title'] = df['title'].apply(get_base_title)

# Identify original: shortest title per artist + main_title
originals = (
    df.sort_values('title', key=lambda x: x.str.len())
      .drop_duplicates(subset=['main_title', 'artist'], keep='first')
      .rename(columns={'title': 'original_title', 'rank': 'original_rank'})
)

# Find highest rank remix
alternates = (
    df[~df.index.isin(originals.index)]
    .sort_values('rank', ascending=False)  # higher rank = more popular
    .drop_duplicates(subset=['main_title', 'artist'], keep='first')
    .rename(columns={'title': 'remix_title', 'rank': 'remix_rank'})
)

# Merge pairs
paired = pd.merge(
    originals[['main_title', 'artist', 'original_title', 'original_rank']],
    alternates[['main_title', 'artist', 'remix_title', 'remix_rank']],
    on=['main_title', 'artist']
)

paired.shape
paired.to_csv('../../Data/Processed/paired_deezer_remixes.csv')
```

```python
# Plot settings
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Layered histogram
sns.histplot(paired['original_rank'], 
             color='skyblue', 
             label='Original Songs', 
             bins=50, 
             kde=False, 
             stat='count',
             element='step', 
             fill=True,
             alpha=0.5)

sns.histplot(paired['remix_rank'], color='lightcoral', label='Remix Songs', bins=50, kde=False, stat='count', element='step', fill=True, alpha=0.5)

# Add log scale if desired
# plt.xscale('log')  # Optional – helps spread out higher ranks if you need it
plt.xscale('log')
# Labels and title
plt.title('Original Songs vs Remix Popularity Rank', fontsize=14)
plt.xlabel('Popularity Rank')
plt.ylabel('Count')
plt.legend(title='Version')
plt.tight_layout()
plt.show()
```
