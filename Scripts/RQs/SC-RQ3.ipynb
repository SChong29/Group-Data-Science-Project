# Research Question 3: What factors influence an artist's success?
Here we want to see if artists should focus more on releasing singles or more albums (comparing popularity scores), how many more streams do they get when they are featured in a song or have an artist(s) featuring in their songs compared to the streams they get when they're in a song on their own, and are they more popular on Spotify or Deezer?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
```

### Subquestion: Solo, leading or featuring - What role produces the most streams?

First we want to see if there's a difference in streams when artists are solo/leading/featured so we load the data containing Spotify's top 3000 artists.

```python
# Read in CSV file of top Spotify artists
df = pd.read_csv('../../Data/Processed/kworb_spotify_artists.csv')
df.columns = df.columns.str.lower()
```

Some artists have never collaborated on a song as a lead artist or featured artist so we want to filter them out by not including any '0' entries.

```python
df = df[(df['tracks_as_lead'] > 0) & (df['tracks_as_feature'] > 0)]
```

Here we create a barchart to plot the mean streams for each of the roles.

```python
# Calculate streams per track for each role
df['streams_per_solo_track'] = df['solo_streams'] / df['tracks_solo']
df['streams_per_lead_track'] = df['streams_as_lead'] / df['tracks_as_lead']
df['streams_per_feature_track'] = df['featured_streams'] / df['tracks_as_feature']

roles = ['Solo', 'Lead', 'Featured']
streams_per_track = [
    df['streams_per_solo_track'].mean(),
    df['streams_per_lead_track'].mean(),
    df['streams_per_feature_track'].mean()
]

# Calculate total tracks for each role
track_counts = [
    df['tracks_solo'].sum(),
    df['tracks_as_lead'].sum(),
    df['tracks_as_feature'].sum()
]
sns.set(style="white")

plt.figure(figsize=(8, 6))
bars = plt.bar(roles, streams_per_track, color='mediumseagreen')

# Annotate bars with number of tracks
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 0.01 * max(streams_per_track),
        f"{track_counts[i]:,} tracks",
        ha='center', va='bottom', fontsize=10, color='black'
    )

plt.title('Streams Per Track for Each Role', fontsize=14)
plt.xlabel('Role', fontsize=12)
plt.ylabel('Mean Streams Per Track', fontsize=12)

# Remove scientific notation from y-axis
ax = plt.gca()
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))

plt.grid(False)

plt.tight_layout()
#plt.savefig('s5_3_streams_per_role.png', dpi=300)
plt.show()
```

Featured tracks have the highest mean streams per track with approx. 3.5 million streams per track
<br>Lead tracks come next, with around 2.7 million streams per track.
<br>Solo tracks have the lowest mean streams per track, at around 2.3 million.

It seems as though featured tracks consistently outperform others in mean streams per track despite the smaller track count. This possibly indicates that being a featured artist can significantly boost an artist's visibility and overall streams.

### Subquestion: Should artists focus more on making singles or albums?
Another factor we want to look at is whether focusing on singles rather than full albums leads to greater success and audience engagement, we will do this by finding the mean popularity score (called rank on Deezer) for both singles and songs in albums.

```python
df = pd.read_csv('../../Data/Processed/spotify_deezer_merged.csv')
```

```python
# Take only relevant fields
df = df[['spotify.name', 
         'spotify.album.name', 
         'spotify.popularity', 
         'spotify.album.total_tracks', 
         'deezer.artist', 
         'deezer.rank']]
df.head()
```

Below we rename the columns to simplify them. <br>
The `spotify.album.total_tracks` column shows how many tracks there are in an album but if the song is a single the amount is one so to make it easier to visualise we'll create a new column called `release_type`.

```python
# First rename columns
df.columns = ['song_name', 'album_name', 'spotify_popularity', 'total_tracks', 'artist', 'deezer_rank']

# Then create the new column using the renamed one
df['track_type'] = df['total_tracks'].apply(lambda x: 'single' if x == 1 else 'album')

#df.to_csv('../../Data/Processed/singles_albums_merged.csv', index=False)
```

Now we make two bar charts to compare the mean popularity scores for singles and albums on Spotify (in green) and the mean popualrity rank (same as popularity score) for singles and albums on Deezer (in purple).

```python
# Group by 'track_type' and calculate the mean for Spotify popularity
spotify_means = df.groupby('track_type')['spotify_popularity'].mean().reset_index()

# Group by 'track_type' and calculate the mean for Deezer rank
deezer_means = df.groupby('track_type')['deezer_rank'].mean().reset_index()

sns.set(style="white")
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Spotify bar chart
sns.barplot(data=spotify_means, x='track_type', y='spotify_popularity', color='mediumseagreen', ax=axes[0])
axes[0].set_title('Mean Spotify Popularity Score: Singles vs Albums')
axes[0].set_xlabel('Track Type')
axes[0].set_ylabel('Mean Popularity Score')

# Deezer bar chart
sns.barplot(data=deezer_means, x='track_type', y='deezer_rank', color='rebeccapurple', ax=axes[1])
axes[1].set_title('Mean Deezer Popularity Score: Singles vs Albums')
axes[1].set_xlabel('Track Type')
axes[1].set_ylabel('Mean Popularity Score')

plt.tight_layout()
#plt.savefig('s5_3_track_type_popularity.png', dpi=300)
plt.show()
```

From the bar charts, we can see that songs in albums typically have better scores than singles on both Spotify and Deezer. This is to be expected because albums often benefit from more extensive promotion and that can lead to higher listener commitment. Artists and labels typically invest more in marketing full albums compared to singles, which often means better visibility on streaming platforms which results in better chart placements and overall song popularity.

We wanted to take a look at the difference between singles and albums from another perspective, so we found the top 5 artists on Spotify and Deezer by taking the highest average popularity scores.
Then we make a filtered df to store the data of the top artists and identify the most popular track type (single or album) for each artist on both platforms.
We group their tracks by type and calculate the mean popularity scores which will determine which track type tends to be more successful for each artist.<br> This gives us insight into whether top-performing artists lean more heavily on releasing singles or albums to achieve their success.

```python
# Filter out artists with fewer than 3 tracks to avoid outliers/bias
artist_counts = df['artist'].value_counts()
valid_artists = artist_counts[artist_counts >= 3].index
filtered_df = df[df['artist'].isin(valid_artists)]

# Calculate top 5 artists by mean Spotify popularity and Deezer rank
spotify_top5 = (
    filtered_df.groupby('artist')['spotify_popularity']
    .mean()
    .nlargest(5)
    .reset_index()
)
spotify_top5 = spotify_top5.merge(
    df[['artist', 'track_type']],
    on='artist',
    how='left'
).drop_duplicates(subset='artist')

deezer_top5 = (
    filtered_df.groupby('artist')['deezer_rank']
    .mean()
    .nlargest(5)
    .reset_index()
)
deezer_top5 = deezer_top5.merge(
    df[['artist', 'track_type']],
    on='artist',
    how='left'
).drop_duplicates(subset='artist')

sns.set(style="white")

# Set colours to colour code singles and albums
pastel_colors = sns.color_palette("pastel")
custom_palette = {'single': pastel_colors[0], 'album': pastel_colors[1]}

fig, axes = plt.subplots(2, 1, figsize=(12, 10))

# Spotify bar chart
sns.barplot(
    data=spotify_top5,
    x='spotify_popularity',
    y='artist',
    hue='track_type',
    dodge=False,
    ax=axes[0],
    palette=custom_palette
)
axes[0].set_title('Top 5 Artists by Mean Spotify Popularity Score')
axes[0].set_xlabel('Mean Spotify Popularity Score')
axes[0].set_ylabel('')
axes[0].set_xlim(0, 100)

# Set legend order
handles, labels = axes[0].get_legend_handles_labels()
order = ['album', 'single']
ordered_handles = [handles[labels.index(label)] for label in order]
axes[0].legend(ordered_handles, order, title='Track Type')

# Deezer bar chart
sns.barplot(
    data=deezer_top5,
    x='deezer_rank',
    y='artist',
    hue='track_type',
    dodge=False,
    ax=axes[1],
    palette=custom_palette
)
axes[1].set_title('Top 5 Artists by Mean Deezer Popularity Score')
axes[1].set_xlabel('Mean Deezer Popularity Score')
axes[1].set_ylabel('')
axes[1].set_xlim(0, 100)

# Set legend order
handles, labels = axes[1].get_legend_handles_labels()
ordered_handles = [handles[labels.index(label)] for label in order]
axes[1].legend(ordered_handles, order, title='Track Type')

plt.tight_layout()
#plt.savefig('s5_3_top5_artists_tracktypes.png', dpi=300)
plt.show()
```

The top plot shows the five artists with the highest average Spotify popularity scores, and highlights whether their most popular track type was a single or album. The bottom plot does the same for Deezer.

From the Spotify chart, we see that a mix of singles and albums contribute to top artist success. For example, Chappell Roan and Gigi Perez achieved high popularity through singles, while others like Evanescence and Dream Supplier saw better performance with albums.

It's a little different on Deezer, the chart shows us that four out of the five artists have higher popularity scores with their singles.

Although earlier we saw that, overall, albums tend to perform better than singles, singles still contribute greatly to an artist's popularity. This tells us that artists could benefit from a mixed release strategy or focus more on catering to their audience’s preferences. However, Deezer does show a slight preference towards singles across the top artists despite album tracks being more well rated.

### Subquestion: Does the streaming platform have an impact on the artists performance?
We also want to know if an artist's popularity score is different on Spotify compared to Deezer, below is a histogram that shows the distribution of artists based on their mean popularity score.

```python
# Calculate mean popularity scores per artist
artist_scores = df.groupby('artist')[['spotify_popularity', 'deezer_rank']].mean().reset_index()

sns.set(style="white")
fig, axes = plt.subplots(2, 1, figsize=(12, 10), sharex=False)

# Spotify histogram
sns.histplot(
    artist_scores['spotify_popularity'],
    bins=100,
    binrange=(0, 100),
    color='mediumseagreen',
    kde=True,
    ax=axes[0]
)
axes[0].set_title('Distribution of Mean Spotify Popularity per Artist', fontsize=14)
axes[0].set_xlabel('Mean Spotify Popularity')
axes[0].set_ylabel('Number of Artists')

# Deezer histogram
sns.histplot(
    artist_scores['deezer_rank'],
    bins=100,
    binrange=(0, 100),
    color='rebeccapurple',
    kde=True,
    ax=axes[1]
)
axes[1].set_title('Distribution of Mean Deezer Rank per Artist', fontsize=14)
axes[1].set_xlabel('Mean Deezer Rank')
axes[1].set_ylabel('Number of Artists')

plt.tight_layout()
#plt.savefig('s5_3_platform_popularity.png', dpi=300)
plt.show()
```

Both platforms show a strong skew toward lower-performing artists, Deezer’s distribution is even more concentrated at the lower end whereas Spotify is more concentrated towards the low-mid popularity scores. This suggests that a very small group of artists dominate visibility and rank, while the vast majority fall well below them. <br>

The stark difference between Spotify and Deezer distributions can be explained by differences in platform scale and regional popularity. Spotify has a much larger global user base, giving a wider range of artists the chance to accumulate streams and popularity. Its algorithm-driven discovery also helps emerging artists gain moderate visibility, resulting in a smoother popularity curve. <br>
Deezer, on the other hand, has a smaller, more regionally concentrated audience, especially in Europe. As a result, only a small number of artists tend to achieve high scores, leading to a sharp drop-off in rank distribution. This reflects how platform size and audience reach can significantly affect artist visibility.
Additionally, when an artist becomes popular on Spotify or Deezer, their listeners are often more likely to stick with that one platform for streaming which could decrease their popularity on the other platform.

### Conclusion

Based on the graphs, we can conclude that an artist's performance may be influenced by the role they play in a song. Artists tend to garner more streams when they are featured on tracks, compared to when they are the lead artist or performing solo. This could be due to the broader audience reach that comes with collaborations.

However, when examining an artist's performance in relation to the type of tracks they release, there is no clear evidence that releasing albums over singles impact an artist's success as seen in the graph of the top 5 artists above. The only notable difference is that albums tend to receive slightly higher popularity scores.

When it comes to streaming platforms, artists trying to make a name for themselves might face different challenges depending on where they release their music. On Spotify, there’s a lot of competition in the mid-tier range, with more artists fighting for attention. Deezer, on the other hand, seems to be tougher when it comes to climbing into the top spots, possibly because it’s smaller or less geared toward pushing new talent. That said, if an artist is aiming for global success, Spotify might be the better option due to its large user base and wider international reach. <br>
Overall, it could be the more strategic choice for getting music in front of a bigger audience.
