Comparing song popularity based on region

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
spotify_data = pd.read_csv('../../Data/Processed/spotify_data.csv')
spotify_data.head()
```

```python
spotify_data.count()
```

Looks like the data is very much all complete, apart from 1 song and 28 isrcs, which is a nice surprise

# Basic Graphs

```python
# Popularity distribution
plt.figure(figsize=(10, 6))
# So it is a histogram, even though it is really a bar chart, but this saves me from a lot of formatting
spotify_data['popularity'].hist(bins=100, color='mediumseagreen', edgecolor='black')
plt.title('Popularity Distribution')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

ISRC gives us the country code, which we can assume to use as a region of origin(although likely not exactly reliable)

```python
# Add a column with the first 2 letters of the ISRC -> this is the country code
spotify_data['isrc_prefix'] = spotify_data['external_ids.isrc'].str[:2]

# Drop rows with no ISRC
spotify_data = spotify_data.dropna(subset=['external_ids.isrc'])

spotify_data.head()
```

So what region codes are used the most? apart from the most obvious answer being US

```python
country_code_counts = spotify_data['isrc_prefix'].value_counts()

# 5 most used
most_used = country_code_counts.head(5)
print("5 Most Used Country Codes:")
print(most_used)

# 5 least used
least_used = country_code_counts.tail(5)
print("\n5 Least Used Country Codes:")
print(least_used)
```

The 5 least used ones are useless unfortunately, but might be worth looking at some with only a few hunderet to compare

```python
# Capitalize all ISRC codes
spotify_data['isrc_prefix'] = spotify_data['isrc_prefix'].str.upper()

# Map QZ, QT, and QM to US in the isrc_prefix column
# All of these are just alternative codes for US
spotify_data['isrc_prefix'] = spotify_data['isrc_prefix'].replace({'QZ': 'US', 'QT': 'US', 'QM': 'US'})

spotify_data['isrc_prefix'].value_counts().head()
```

Before that, the fact that GB and US are separated even though they speak the same language(together with some other countries like IE or Australia) should mean that the only popularity difference should come from how good the artists are

```python
us_color = 'black'
gb_color = 'blue'

us_data = spotify_data[spotify_data['isrc_prefix'] == 'US']['popularity']
gb_data = spotify_data[spotify_data['isrc_prefix'] == 'GB']['popularity']

# Plot the histograms with percentages
plt.figure(figsize=(10, 6))
plt.hist(us_data, bins=100, alpha=1, label='US', color=us_color, edgecolor=None, density=True, range=(0, 100))
plt.hist(gb_data, bins=100, alpha=0.5, label='GB', color=gb_color, edgecolor=None, density=True, range=(0, 100))
plt.title('Popularity Distribution for US and GB')
plt.xlabel('Popularity')
plt.ylabel('Percentage')

# Add vertical lines for average popularity
us_avg = us_data.mean()
gb_avg = gb_data.mean()
plt.axvline(us_avg, color=us_color, linestyle='--', linewidth=1, label=f'US Avg: {us_avg:.1f}')
plt.axvline(gb_avg, color=gb_color, linestyle='--', linewidth=1, label=f'GB Avg: {gb_avg:.1f}')

plt.legend()
# plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

Not really any difference here, what could be some more/less popular regions?

So what regions are the most/least popular?

```python
average_popularity_by_region = spotify_data.groupby('isrc_prefix')['popularity'].mean()

# Top 3
top_3_regions = average_popularity_by_region.nlargest(3)
print("Top 3 Regions with Highest Average Popularity:")
for region, avg_popularity in top_3_regions.items():
    song_count = country_code_counts.get(region, 0)
    print(f"Region: {region}, Average Popularity: {avg_popularity:.2f}, Song Count: {song_count}")

# Bottom 3
bottom_3_regions = average_popularity_by_region.nsmallest(3)
print("\nBottom 3 Regions with Lowest Average Popularity:")
for region, avg_popularity in bottom_3_regions.items():
    song_count = country_code_counts.get(region, 0)
    print(f"Region: {region}, Average Popularity: {avg_popularity:.2f}, Song Count: {song_count}")
```

Clearly 1 song is not exactly indicative, so lets do 100 or more songs to see what happens:

```python
# Regions with 100 or more songs
filtered_regions = average_popularity_by_region[country_code_counts >= 100]

# Top 3
top_3_regions = filtered_regions.nlargest(3)
print("Top 3 Regions with Highest Average Popularity (with at least 100 songs):")
for region, avg_popularity in top_3_regions.items():
    song_count = country_code_counts.get(region, 0)
    print(f"Region: {region}, Average Popularity: {avg_popularity:.2f}, Song Count: {song_count}")

# Bottom 3
bottom_3_regions = filtered_regions.nsmallest(3)
print("\nBottom 3 Regions with Lowest Average Popularity (with at least 100 songs):")
for region, avg_popularity in bottom_3_regions.items():
    song_count = country_code_counts.get(region, 0)
    print(f"Region: {region}, Average Popularity: {avg_popularity:.2f}, Song Count: {song_count}")
```

That is a bit more interesing, looks like there are more pople listening to the few songs from India and Mexico than to the huge amount of songs from US. There is also the curious case of Iceland(IS) where it looks like they have too many songs for the size of population and speakers.

Lets take a look at US and IN comparison first before looking at a scatter plot of this

```python
country1 = 'IN'
country2 = 'US'

us_color = 'black'
gb_color = 'blue'

us_data = spotify_data[spotify_data['isrc_prefix'] == country1]['popularity']
gb_data = spotify_data[spotify_data['isrc_prefix'] == country2]['popularity']

plt.figure(figsize=(10, 6))
# Reduced bins to 50 for less spikeness, as IN doesn't have as many songs
plt.hist(us_data, bins=50, alpha=1, label=country1, color=us_color, edgecolor=None, density=True, range=(0, 100))
plt.hist(gb_data, bins=50, alpha=0.6, label=country2, color=gb_color, edgecolor=None, density=True, range=(0, 100))
plt.title(f'Popularity Distribution for {country1} and {country2}')
plt.xlabel('Popularity')
plt.ylabel('Percentage')

# Vertical lines for average popularity
us_avg = us_data.mean()
gb_avg = gb_data.mean()
plt.axvline(us_avg, color=us_color, linestyle='--', linewidth=1, label=f'{country1} Avg: {us_avg:.1f}')
plt.axvline(gb_avg, color=gb_color, linestyle='--', linewidth=1, label=f'{country2} Avg: {gb_avg:.1f}')

plt.legend()
# plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

Looks like IN has a bigger spike in the 60-80 interval, but US still beats it in the most popular songs(Likely a language barrier difference)

## Graphs based on year of release

```python
"""
# Extract the year from the ISRC (characters 6-7 represent the year in ISRC)
spotify_data['release_year'] = spotify_data['external_ids.isrc'].str[5:7]

# Convert release_year to numeric, coercing errors to NaN, then drop non-numeric rows
# So what happened here is that some isrc codes gave not a number in those spots and I do not know why
spotify_data['release_year'] = pd.to_numeric(spotify_data['release_year'], errors='coerce')
spotify_data = spotify_data.dropna(subset=['release_year'])
spotify_data['release_year'] = spotify_data['release_year'].astype(int)
"""

# Extract the year from album.release_date
spotify_data['release_year'] = spotify_data['album.release_date'].str[:4].astype(int)

spotify_data.head()
```

Song count per year

```python
song_count_by_year = spotify_data['release_year'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.hist(song_count_by_year.index, bins=126, weights=song_count_by_year.values, color='mediumseagreen', alpha=0.7, edgecolor='black', range=(1900, 2025))
plt.title('Song Count by Year')
plt.xlabel('Year')
plt.ylabel('Song Count')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

Turns out that 2014 is when kworb started collecting top charts, sso lets split data to before and after!

```python
filtered_data = spotify_data[spotify_data['release_year'] > 2014]

plt.figure(figsize=(12, 6))
plt.title('Song Popularity by Year (From 2015)')
plt.xlabel('Year')
plt.ylabel('Average Popularity')
plt.grid(axis='y', linestyle='-', alpha=0.5)

# Regression line
m, b = np.polyfit(filtered_data['release_year'], filtered_data['popularity'], 1)
# MSE
predicted_popularity = m * filtered_data['release_year'] + b
mse = np.mean((filtered_data['popularity'] - predicted_popularity) ** 2)
# Scatterplot for popularity by year
plt.scatter(filtered_data['release_year'], filtered_data['popularity'], color='mediumseagreen', alpha=0.002, edgecolor=None, label='Popularity Data Points')
plt.plot(filtered_data['release_year'], predicted_popularity, color='red', linestyle='-', label='Regression Line (MSE: {:.2f})'.format(mse))

plt.legend()
plt.show()
```

Newer songs perform a bit better, likely because they are new, what about older songs?

```python
filtered_data = spotify_data[(spotify_data['release_year'] < 2013) & (spotify_data['release_year'] >= 1980)]

plt.figure(figsize=(12, 6))
plt.title('Song Popularity by Year (until 2013)')
plt.xlabel('Year')
plt.ylabel('Average Popularity')
plt.grid(axis='y', linestyle='-', alpha=0.5)

# Regression line
m, b = np.polyfit(filtered_data['release_year'], filtered_data['popularity'], 1)
# MSE
predicted_popularity = m * filtered_data['release_year'] + b
mse = np.mean((filtered_data['popularity'] - predicted_popularity) ** 2)
# Scatterplot for popularity by year
plt.scatter(filtered_data['release_year'], filtered_data['popularity'], color='mediumseagreen', alpha=0.1, edgecolor=None, label='Popularity Data Points')
plt.plot(filtered_data['release_year'], predicted_popularity, color='red', linestyle='-', label='Regression Line (MSE: {:.2f})'.format(mse))

plt.legend()
plt.show()
```

The smallest value being in 2010 is interesting, but the general down trend could be atributed to less songs making it on the charts, but when they did, they had already proven themselves to be good enough to stay popular

Song Length stuff? I don't think there is anything here

```python
# Calculate the 25th and 75th percentiles for song length and popularity
duration_25th = spotify_data['duration_ms'].quantile(0.05)
duration_75th = spotify_data['duration_ms'].quantile(0.95)

# Filter the data
filtered_data = spotify_data[
    (spotify_data['duration_ms'] >= duration_25th) & 
    (spotify_data['duration_ms'] <= duration_75th)
]

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(filtered_data['duration_ms'] / 1000, filtered_data['popularity'], alpha=0.01, color='mediumseagreen', edgecolor=None)
plt.title('Popularity vs Song Length')
plt.xlabel('Song Length (seconds)')
plt.ylabel('Popularity')
plt.grid(True, linestyle='--', alpha=0.5)

# Fit a polynomial curve to the data
z = np.polyfit(filtered_data['duration_ms'] / 1000, filtered_data['popularity'], 8)
p = np.poly1d(z)

# Generate x values for the curve
x = np.linspace(filtered_data['duration_ms'].min() / 1000, filtered_data['duration_ms'].max() / 1000, 500)
y = p(x)

# Draw a curve along the top points
top_points = filtered_data.groupby(filtered_data['duration_ms'] // 1000)['popularity'].max().reset_index()
z_top = np.polyfit(top_points['duration_ms'], top_points['popularity'], 8)
p_top = np.poly1d(z_top)

# Generate x values for the top curve
x_top = np.linspace(top_points['duration_ms'].min(), top_points['duration_ms'].max(), 500)
y_top = p_top(x_top)

# Plot the curves
plt.plot(x_top, y_top, color='blue', linestyle='--', label='Top Points Curve')
plt.plot(x, y, color='red', linestyle='-', label='Fitted Curve')

plt.legend()

plt.show()
```

Most of the songs lay in the 2-4 minute range, but popularity average stays around the same regardless of how long a song is
