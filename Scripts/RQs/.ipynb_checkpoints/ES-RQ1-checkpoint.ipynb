# RQ1:Are remixes or alternate versions of songs more likely to be popular on Deezer compared to Spotify?


## Popularity Trends Every 5 years (Deezer and Spotify)

This analyzes how the popularity scores of **Original** vs **Remix/Alternate** tracks has changed over time on both platforms.

- The merged dataset (`spotify_deezer_merged.csv`) is loaded from the `../../Data/Processed` directory.
- A `track_type` column is created to classify tracks as either *Original* or *Remix/Alternate* based on the keyword "remix".
- The dataset is filtered to include only the years: 2000, 2005, 2010, 2015, 2020, and 2025.
- Two bar charts are shown:
  - **Deezer Rank (popularity score) by Year and Track Type**
  - **Spotify Popularity by Year and Track Type**

These plots help us see how remix visibility has developed compared to original tracks across years on each of these platforms.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

PROCESSED_DIR = os.path.abspath(os.path.join(os.getcwd(), "../../Data/Processed"))
merged_file = os.path.join(PROCESSED_DIR, "spotify_deezer_merged.csv")

# Load datset
merged_df = pd.read_csv(merged_file)

merged_df['year'] = pd.to_datetime(
    merged_df['spotify.album.release_date'], errors='coerce'
).dt.year.astype('Int64')

# Add 'track_type' column based on title keywords 
merged_df['track_type'] = merged_df['deezer.title'].str.lower().apply(
    lambda x: 'Remix/Alternate' if 'remix' in x or 'alternate' in x else 'Original'
)

# Filter for 5-year intervals from 2000 to 2025 inclusive
# (2000, 2005, 2010, 2015, 2020, 2025)
years = list(range(2000, 2026, 5))
filtered_df = merged_df[merged_df['year'].isin(years)]

sns.set(style='whitegrid')
palette = "Set2"

# Deezer popularity 
plt.figure(figsize=(12, 6))
sns.barplot(data=filtered_df, x="year", y="deezer.rank", hue="track_type", palette=palette, ci=None)
plt.title("Deezer Popularity Score by Year: Original vs Remix/Alternate")
plt.xlabel("Year")
plt.ylabel("Deezer Popularity Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(PROCESSED_DIR, "s5_1_deezer_5_year.png"))
plt.show()
plt.close()

# Spotify popularity  
plt.figure(figsize=(12, 6))
sns.barplot(data=filtered_df, x="year", y="spotify.popularity", hue="track_type", palette=palette, ci=None)
plt.title("Spotify Popularity Score by Year: Original vs Remix/Alternate")
plt.xlabel("Year")
plt.ylabel("Spotify Popularity Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(PROCESSED_DIR, "s5_1_spotify_5_year.png"))
plt.show()
plt.close()
```

## T - test Comparison: Original vs Remix Popularity

This part does a statistical comparison between **Original** and **Remix/Alternate** tracks on  Deezer and Spotify.

- Deezer ranks are **normalized** to a 01 scale to match Spotify's popularity score.
- A `track_type` column is created to classify tracks as either *Original* or *Remix/Alternate* based on the keyword "remix".
- Two **T-tests** were generated:
  - One for Deezer popularity scores (rank)
  - One for Spotify popularity scores
- The results help us see if the differences in popularity are statistically significant.
- Boxplots are created for each platform to visualize the distribution of popularity scores by track type.

This analysis goal is to answer whether remixes are more popular than original versions.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import os

PROCESSED_DIR = os.path.abspath(os.path.join(os.getcwd(), "../../Data/Processed"))
merged_file = os.path.join(PROCESSED_DIR, "spotify_deezer_merged.csv")

print(f"Reading file from: {merged_file}")

# Load dataset
merged_df = pd.read_csv(merged_file)

# Normalize Deezer rank to the range 0 to 1
merged_df['deezer_popularity_score'] = (
    (merged_df['deezer.rank'] - merged_df['deezer.rank'].min()) /
    (merged_df['deezer.rank'].max() - merged_df['deezer.rank'].min())
)

# Add 'track_type' column based on title keywords 
merged_df['track_type'] = merged_df['deezer.title'].apply(
    lambda x: 'Remix/Alternate' if 'Remix' in x else 'Original'
)

# T-test for Deezer Popularity Score
original_deezer = merged_df[merged_df['track_type'] == 'Original']['deezer_popularity_score']
remix_deezer = merged_df[merged_df['track_type'] == 'Remix/Alternate']['deezer_popularity_score']
deezer_ttest_stat, deezer_ttest_pvalue = stats.ttest_ind(original_deezer, remix_deezer, nan_policy='omit')

# T-test for Spotify Popularity Score
original_spotify = merged_df[merged_df['track_type'] == 'Original']['spotify.popularity']
remix_spotify = merged_df[merged_df['track_type'] == 'Remix/Alternate']['spotify.popularity']
spotify_ttest_stat, spotify_ttest_pvalue = stats.ttest_ind(original_spotify, remix_spotify, nan_policy='omit')

sns.set(style="whitegrid")

# Deezer Popularity Score Boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=merged_df, x="track_type", y="deezer_popularity_score", palette="Set2")
plt.title("Deezer Popularity Score: Original vs Remix/Alternate")
plt.xlabel("Track Type")
plt.ylabel("Deezer Popularity Score")
plt.savefig("s5_1_deezer_popularity_boxplot.png")
plt.show()

# Spotify Popularity Score Boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=merged_df, x="track_type", y="spotify.popularity", palette="Set2")
plt.title("Spotify Popularity Comparison: Original vs Remix/Alternate")
plt.xlabel("Track Type")
plt.ylabel("Spotify Popularity Score")
plt.savefig("s5_1_spotify_popularity_boxplot.png")
plt.show()

# Print T-test 
print(f"Deezer T-test result:\nStatistic: {deezer_ttest_stat}, P-value: {deezer_ttest_pvalue}")
print(f"Spotify T-test result:\nStatistic: {spotify_ttest_stat}, P-value: {spotify_ttest_pvalue}")
```



### Conclusion:
- **Deezer** shows that **Original** tracks are less popular than **Remix/Alternate** tracks.
- **Spotify** shows that **Remix/Alternate** tracks are less popular than **Original** tracks.

Both results are statistically significant, indicating that track type (Original vs. Remix/Alternate) has a meaningful impact on the popularity scores on both platforms.
